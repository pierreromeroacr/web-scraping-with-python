## urllib.request
es un módulo Python para acceder y utilizar recursos de internet identificados por URLs (Uniform Resource Locators). Ofrece una interfaz muy simple, a través de la función urlopen. Esta función es capaz de acceder a URLs usando una variedad de protocolos diferentes.

```python
import urllib.request

#Abre la URL y guarda el objeto resultante con "urlopen" y lo almacenamos en "openurl"
openurl = urllib.request.urlopen('https://www.w3.org/Summary.txt')

#Itera sobre cada línea del contenido obtenido de la URL
for x in openurl:
    #convierte cada línea a una cadena de texto utilizando el método "decode()",
    #y luego elimina los espacios en blanco alrededor de la cadena utilizando el método "strip()"
    print(x.decode().strip())
```

### Explotando la heramienta BeautifulSoup

```python
import urllib
from bs4 import BeautifulSoup

url = "https://www.w3.org"
openurl = urllib.request.urlopen(url)

#extraer el código HTML de esta página
parse = BeautifulSoup(openurl)

#obtener los links <a>
etiqueta = parse('a')

for x in etiqueta:
    #etiqueta completa
    print('Etiqueta:', x)
    #obteniendo el valor del href de la etiqueta <a>
    print("url:", x.get('href'))
    #obteniendo contenido
    print("contenido:", x.contents)
    #obteniendo atributo
    print("atributo:", x.attrs)
    print('\n')
```

### Analizando el texto extraido de las etiquetas

```python
import urllib
from bs4 import BeautifulSoup

url = "https://peru21.pe/peru/gran-noticia-peru-se-posiciona-como-lider-mundial-de-diversidad-de-aves-especies-nivel-mundial-servicio-nacional-de-areas-naturales-protegidas-sernanp-noticia/"
openurl = urllib.request.urlopen(url)

#extraer el código HTML de esta página
parse = BeautifulSoup(openurl)

#obtener los parrafos
etiqueta = parse('b')

text = ''
#Itera sobre todas las etiquetas de párrafo encontradas y agrega el contenido de cada párrafo a la cadena de texto si la etiqueta no tiene atributos
for x in etiqueta:
    if(len(x.attrs)) == 0:
        a = x.contents[0]
        #concatenando los elementos encontrados
        text = str(text) + str(a)

#Analizar cuantas palabras se repiten en el texto
data_count = dict()

#reemplazando signos por espacios en blanco
text = text.replace(',', '').replace('.', '').replace(':', '').replace('¡', '').replace('!', '').replace('(', '').replace(')', '')

#convirtiendo todo el texto a minuscula y lo convertimos en un lista
words = text.lower().split()

# contar la frecuencia de palabras en lista
for x in words:
    if len(x) > 1:
        data_count[x] = data_count.get(x, 0) +1

print(data_count)
```

```python
#muestra los elementos del diccionario data_count en forma de pares clave-valor
data = data_count.items()
print(data)
```

```python
#mostrar el resultado en un datframe
import pandas as pd

df = pd.DataFrame(data, columns=['words', 'count']).sort_values('count', ascending=False)
df
```
